```python
import os
import json
import pandas
```

```python
idir = "/home/akhilesh/cdl/lgl/pocso/Delhi/"
```

```python
metadata = []
for dirs in os.listdir(idir):
    dirs = idir + dirs
    for idirs in os.listdir(dirs):
        idirs = dirs + "/" + idirs
        for filename in os.listdir(idirs):
            if filename.endswith('.json'):
                with open(idirs + "/" + filename) as f:
                    data = json.load(f)
                    metadata.append(data)
```

```python
df = pandas.DataFrame.from_records(metadata).fillna(0)
```

```python
def filter_by_date(year: int, n: int=None):
    """
    filter by date and the number of judgements in the mentioned timeframe
    """
    if n: 
        filtered_date_df = df[df['date_of_decision_dt'].dt.year == year][:n]
        return filtered_date_df
    return df[df['date_of_decision_dt'].dt.year == year]
```

```python
import os
```

```python
cino = filtered_df['cino'].tolist()
```

```python
all_judgements = []
for ci in cino:
    for district in os.listdir('/home/akhilesh/cdl/lgl/pocso/Delhi/'):
        for repo in os.listdir('/home/akhilesh/cdl/lgl/pocso/Delhi/' + district):
            if ci in repo:
                for judgement in os.listdir('/home/akhilesh/cdl/lgl/pocso/Delhi/' + district + "/" + repo):
                    if judgement.endswith('.pdf'):
                        all_judgements.append(judgement.split('.')[0])
```
